```python
path = Path('/content/dogscats')
bs = 32
```

```python
tfms = get_transforms()
```

```python
data = ImageDataBunch.from_folder(path,train='train', valid="valid",
                                  ds_tfms=tfms, size=224, bs=bs)
data.normalize(imagenet_stats)
```

```python
data.show_batch(rows=3)
```

```python
arch = models.resnet34
learn = create_cnn(data, arch, metrics=accuracy)
```

```python
learn.fit(epochs=3, lr=0.001)
```

```python
learn.recorder.plot_losses()
```

```python
img = open_image("")
```

```python
learn.predict(img)
```
